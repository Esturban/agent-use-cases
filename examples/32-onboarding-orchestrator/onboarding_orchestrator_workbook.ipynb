```python
from dotenv import load_dotenv
load_dotenv()
print('Environment loaded')
```

```python
from src.schema import NewHire
from src.workflow import run

hire = NewHire(
    name='Jane Smith',
    role='Senior Software Engineer',
    department='Engineering',
    start_date='2026-07-14',
    location='New York',
)

plan = run(hire)
print(f'Day 1 Ready: {plan.day1_ready}')
print(f'Blockers: {plan.blockers or "None"}')
print(f'Summary: {plan.summary}')
```

```python
for domain_status in [plan.it_status, plan.hr_status, plan.facilities_status]:
    print(f'\n=== {domain_status.domain} ===')
    print('Completed before Day 1:')
    for t in domain_status.completed:
        print(f'  [x] {t}')
    print('Pending (requires employee presence):')
    for t in domain_status.pending:
        print(f'  [ ] {t}')
    if domain_status.notes:
        print(f'Notes: {domain_status.notes}')
```
