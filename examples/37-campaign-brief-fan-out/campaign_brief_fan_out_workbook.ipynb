# 37 — Campaign Brief Fan-Out

One campaign brief fans out in parallel to email, social, and blog specialists, assembled into a typed ContentPack.

```python
from dotenv import load_dotenv
load_dotenv()
print('Environment loaded')
```

```python
from src.schema import CampaignBrief
from src.workflow import run

brief = CampaignBrief(
    topic='Launch of TaskFlow Pro — AI-powered project management',
    target_audience='Startup founders and team leads at 10-100 person companies overwhelmed by project chaos',
    key_message='TaskFlow Pro uses AI to automatically prioritise your team work so you ship faster with less stress',
    tone='casual',
    call_to_action='Start your free 14-day trial at taskflow.io',
)

pack = run(brief)
print(f'Content pack ready for: {pack.topic}')
print(f'Email subject: {pack.email.subject}')
print(f'Blog title: {pack.blog.title}')
```

```python
print('=== EMAIL ===')
print(f'Subject: {pack.email.subject}')
print(f'Preview: {pack.email.preview_text}')
print(f'CTA: {pack.email.cta_button_text}')
print()
print('=== SOCIAL POSTS ===')
for post in pack.social:
    print(f'[{post.platform.upper()}] {post.caption}')
    print(f'  Hashtags: {" ".join("#" + h for h in post.hashtags)}')
    print()
print('=== BLOG OUTLINE ===')
print(f'Title: {pack.blog.title}')
print(f'Meta: {pack.blog.meta_description}')
print(f'~{pack.blog.estimated_word_count} words')
for i, section in enumerate(pack.blog.sections, 1):
    print(f'  {i}. {section}')
```
