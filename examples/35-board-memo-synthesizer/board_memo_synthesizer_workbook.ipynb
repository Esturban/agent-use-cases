```python
from src.workflow import run

topic = 'Proposed acquisition of NovaTech Inc. by MegaCorp Ltd.'
reports = [
    'Goldman Sachs: NovaTech shows 45% YoY ARR growth, 122% NRR, strong management team. Fair at 8x revenue.',
    'Morgan Stanley: Customer concentration risk high (38% from top 3). Integration 18-24 months. 35x EBITDA leaves no margin.',
    'Deloitte Risk: Regulatory approval 9-12 months, antitrust risk. Key-person risk: CEO/CTO retention only 12 months. EU data sovereignty issues.',
]

memo = run(topic, reports)
print(f'POSITION: {memo.recommended_position.upper()}')
print(f'Verdict: {memo.one_sentence_verdict}')
```

```python
print('=== Executive Summary ===')
print(memo.executive_summary)
print()
for opinion in [memo.bull_case, memo.bear_case, memo.risk_case]:
    print(f'--- {opinion.lens.upper()} CASE (confidence: {opinion.confidence}) ---')
    for pt in opinion.key_points:
        print(f'  * {pt}')
    print()
```
