Queries ClinicalTrials.gov (no API key needed) and ranks trials by patient eligibility fit.

```python
from dotenv import load_dotenv
load_dotenv()
print('Environment loaded')
```

```python
from src.schema import PatientCriteria
from src.workflow import run

criteria = PatientCriteria(
    condition='type 2 diabetes',
    age=55,
    location='United States',
    exclusions=['insulin dependent', 'severe kidney disease'],
)

result = run(criteria)
print(f'Condition: {result.condition}')
print(f'Trials found: {result.trials_found}')
print(f'Eligible matches returned: {len(result.matches)}')
```

```python
for match in result.matches:
    conf_icon = {'high': 'H', 'medium': 'M', 'low': 'L'}.get(match.match_confidence, '?')
    print(f'[{conf_icon}] {match.nct_id}: {match.title}')
    print(f'    Phase: {match.phase} | Status: {match.status}')
    print(f'    {match.eligibility_summary}')
    print(f'    Match reason: {match.why_matches}')
    print()
```
