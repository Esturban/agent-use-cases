```python
from dotenv import load_dotenv
load_dotenv()
print('Environment loaded')
```

```python
from src.workflow import run

digest = run('Apple', days_back=7)
print(f'Brand: {digest.brand}')
print(f'Articles analysed: {digest.articles_analysed}')
print(f'Overall sentiment: {digest.overall_sentiment} (avg score: {digest.average_score:.2f})')
print(f'Trend: {digest.trend_direction}')
print(f'\nWeek summary:\n{digest.week_summary}')
```

```python
print(f'Positive: {digest.positive_count} | Negative: {digest.negative_count} | Neutral: {digest.neutral_count}')
print('\nNotable shifts:')
for shift in digest.notable_shifts:
    print(f'  - {shift}')

print('\nPer-article breakdown:')
for article in digest.articles:
    icon = '+' if article.sentiment == 'positive' else ('-' if article.sentiment == 'negative' else '=')
    print(f'  [{icon}] {article.headline[:80]} ({article.score:+.2f})')
```
